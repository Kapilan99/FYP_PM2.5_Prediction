```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
path = ['03_Kandy_hourly.csv', '30_Mannar_hourly.csv', '31_Jaffna_hourly.csv', 
        '32_PointPedro_hourly.csv', '33_Kilinochchi_hourly.csv', '34_Mullativu_hourly.csv', 
        '42_Anuradhapura_hourly.csv']

num = len(path)

for i in range(num):  # Corrected the loop syntax
    file_path = path[i]
    data = pd.read_csv(file_path)

    # Print the column names to inspect
    print(f"Columns in {file_path}: {data.columns}")

    # Convert timestamp to datetime format for proper plotting
    data['timestamp_index'] = pd.to_datetime(data['timestamp_index'])

    # Updated PM column name to 'PM25'
    pm_column = 'PM25'  # Corrected column name

    if pm_column in data.columns:
        # Plot hourly PM2.5 concentration
        plt.figure(figsize=(12, 6))
        plt.plot(data['timestamp_index'], data[pm_column], label='PM2.5 Concentration', color='blue', linewidth=1)

        # Customize the plot
        plt.title(f'Hourly PM2.5 Concentration - {file_path}', fontsize=16)  # Title includes filename
        plt.xlabel('Timestamp', fontsize=12)
        plt.ylabel('PM2.5 (µg/m³)', fontsize=12)
        plt.grid(True, linestyle='--', alpha=0.6)
        plt.legend(fontsize=12)
        plt.tight_layout()

        # Show the plot
        plt.show()
    else:
        print(f"PM2.5 column not found in {file_path}. Please check the column name.")
```
